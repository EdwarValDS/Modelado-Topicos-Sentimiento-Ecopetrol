# Requirements

```python
import pandas as pd
import os
from deep_translator import GoogleTranslator
```

# Data

```python
data = pd.read_csv("./Datos/Datos_analisis_EC_traducidos.csv")
data = data[["Date","Title","Link","Headline","Article","Source"]]

data_to_translate = data.loc[data["Source"]=="La República"]
data_to_not_translate = data.loc[data["Source"]=="Hydrocarbons"]
data_to_translate['Article'] = data_to_translate['Article'].apply(lambda x: x.split('. ') if isinstance(x, str) else None)
data_to_translate = data_to_translate.explode('Article')
```

# Translation

```python
num_splits = 10
split_size = len(data_to_translate) // num_splits
dataframes = []

# Cambiar rango en base a los DF faltantes:
for i in range(num_splits):
    if i < num_splits - 1:
        dataframes.append(data_to_translate.iloc[i * split_size:(i + 1) * split_size])
    else:
        dataframes.append(data_to_translate.iloc[i * split_size:])
```

```python
# Definir las columnas que se van a traducir
columns_to_translate = ["Article"]

translated_dataframes = []

for idx, df in enumerate(dataframes):
    for column in columns_to_translate:
        try:
          df[column] = df[column].apply(lambda x: GoogleTranslator(source='es', target='en').translate(x))
        except:
          pass
    translated_dataframes.append(df)
    df.to_csv(f"./Traducciones/translation_{idx}.csv")
```
